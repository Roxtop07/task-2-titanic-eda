```python
import pandas as pd

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

# -------------------------------
# üìä Summary Statistics & Insights
# -------------------------------

print("üîé DATA STRUCTURE:\n")
df.info()
print("\n‚ùì MISSING VALUES (Column-wise):")
missing = df.isnull().sum()
missing = missing[missing > 0]  # filter out columns with 0 missing

if not missing.empty:
    print(missing)
else:
    print("No missing values in the dataset ‚úÖ")


print("\nüìä DESCRIPTIVE STATISTICS (Numerical Columns):")
print(df.describe())

print("\nüî§ CATEGORICAL VALUE COUNTS:")

# Print value counts for important categorical features
categorical_cols = ['Sex', 'Embarked', 'Pclass']
for col in categorical_cols:
    print(f"\n‚ñ∂Ô∏è {col} value counts:")
    print(df[col].value_counts())
```

```python

# Histograms for all numeric features
df.hist(bins=20, figsize=(12,8), color='steelblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.tight_layout()
plt.savefig("histograms.png")
plt.show()
```

```python

# Boxplots for selected numeric features
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Age', 'Fare']], palette="coolwarm")
plt.title("Boxplots of Age and Fare")
plt.savefig("boxplots.png")
plt.show()
```

```python

# Pairplot to see feature interactions
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.savefig("pairplot.png")
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset (if not already loaded)
df = pd.read_csv("Titanic-Dataset.csv")  # Or use df if already in memory

# Optional: Drop non-numeric columns before correlation
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, linecolor='white')
plt.title("Correlation Heatmap", fontsize=16)
plt.tight_layout()
plt.savefig("correlation-heatmap.png")
plt.show()

```


## üìù Observations

- **Missing Values:** Age has some missing data; Cabin likely dropped.
- **Distributions:** Fare and Age are skewed.
- **Correlations:**
  - `Fare` shows some positive correlation with `Survived`
  - `Pclass` is negatively correlated with `Survived`
- **Outliers:** Detected in both Age and Fare via boxplots.
- **Pairplot:** Shows some separation of survival status based on Fare and Pclass.
